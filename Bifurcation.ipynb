## This code plots the bifurcation in a dynamical system.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
def logistic(r, x):
    """Implementation of the logistic map
    
    Args:
        x (float): previous value from 0 to 1
        r (float): the R paremter
    """
    return r * x * (1 - x)
```

## Plot

```python
n = 1000000  # total number of points
r = np.linspace(0.0, 4.0, n)
iterations = 100
```

```python
x = np.random.uniform(0.0, 1.0, n)  # initial value
for _ in range(iterations):  # iterate
    x = logistic(r, x)

fig = plt.figure(figsize=(20, 10))
ax = plt.axes()

ax.set_title(f"Bifurcation Diagram for $x^{{(r)}}_{{n + 1}} = rx^{{(r)}}_{{n}}(1 - x^{{(r)}}_{{n}})$ for $n=${iterations}", fontsize=24)
ax.set_xlabel('r', fontsize=24)
ax.set_ylabel('$x^{(r)}_{n}$', fontsize=24)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_xlim(0.0, 4.0)
ax.set_ylim(0.0, 1.0)
plt.gcf().text(0.13, 0.15, 'Bifurcation Diagram', fontsize=22, fontfamily='Verdana')

ax.plot(r, x, ',b', alpha=.5)
plt.savefig('bifurcation.png', bbox_inches='tight')
```

## Make animation.

```python
n = 1000000  # total number of points
r = np.linspace(0.0, 4.0, n)
iterations = 50
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes()

def animate(i):
    # clear axes object
    plt.cla()
    ax.set_title(f"Bifurcation Diagram for $x^{{(r)}}_{{n + 1}} = rx^{{(r)}}_{{n}}(1 - x^{{(r)}}_{{n}})$ for $n=${i + 1}", fontsize=24)
    ax.set_xlabel('r', fontsize=24)
    ax.set_ylabel('$x^{(r)}_{n}$', fontsize=24)
    ax.tick_params(axis='both', which='major', labelsize=18)
    ax.set_xlim(0.0, 4.0)
    ax.set_ylim(0.0, 1.0)
    plt.gcf().text(0.13, 0.15, 'Bifurcation Diagram', fontsize=22, fontfamily='Verdana')
    
    x = np.random.uniform(0.0, 1.0, n)  # initial value
    for _ in range(i + 1):  # animated iterations
        x = logistic(r, x)
    
    l = ax.plot(r, x, ',b', alpha=.5)
    return l

# call the animator	 
anim = animation.FuncAnimation(fig, animate, frames=iterations, interval=225, blit=True)
# save the animation as a gif file 
anim.save('biffurcation_rand_init_value.gif',writer='imagemagick') 
```
